```python
import numpy as np
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import  Dense, Flatten, Activation
from keras.utils import to_categorical
from keras.optimizers import SGD
from keras.utils import np_utils
import matplotlib.pyplot as plt
from keras import optimizers
from keras.callbacks import LambdaCallback
```

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

# normalization
x_train = X_train.reshape(60000,784)/255
x_test = X_test.reshape(10000,784)/255
```

```python
# model 1
model = Sequential()
model.add(Dense(input_dim=28*28, units=128))
model.add(Activation('sigmoid'))
model.add(Dense( units=10))
model.add(Activation('softmax'))

weights1 = []
print_weights1 = LambdaCallback(on_epoch_end=lambda batch, logs: weights1.append(model.get_weights()))

# compile the model 
sgd = optimizers.SGD(lr=0.01, momentum=0.0, decay=0.0, nesterov=False)
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])

# training
history1=model.fit(
  x_train,
  to_categorical(Y_train),
  batch_size=100,
  epochs=40,
  validation_data=(x_test, to_categorical(Y_test)),
  callbacks = [print_weights1]
)
```

```python
# model 2
model = Sequential()
model.add(Dense(input_dim=28*28, units=128))
model.add(Activation('sigmoid'))
model.add(Dense( units=10))
model.add(Activation('softmax'))

weights2 = []
print_weights2 = LambdaCallback(on_epoch_end=lambda batch, logs: weights2.append(model.get_weights()))

# compile the model 
sgd = optimizers.SGD(lr=0.1, momentum=0.0, decay=0.0, nesterov=False)
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])

# training
history2=model.fit(
  x_train,
  to_categorical(Y_train),
  batch_size=100,
  epochs=40,
  validation_data=(x_test, to_categorical(Y_test)),
  callbacks = [print_weights2]
)
```

```python
# model 3
model = Sequential()
model.add(Dense(input_dim=28*28, units=128))
model.add(Activation('relu'))
model.add(Dense( units=10))
model.add(Activation('softmax'))

weights3 = []
print_weights3 = LambdaCallback(on_epoch_end=lambda batch, logs: weights3.append(model.get_weights()))

# compile the model 
sgd = optimizers.SGD(lr=0.01, momentum=0.0, decay=0.0, nesterov=False)
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])

# training
history3=model.fit(
  x_train,
  to_categorical(Y_train),
  batch_size=100,
  epochs=40,
  validation_data=(x_test, to_categorical(Y_test)),
  callbacks = [print_weights3]
)
```
